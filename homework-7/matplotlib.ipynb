```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import scipy.special
```

## Exercise 1  
1. Создайте график $x^2⋅sin(1/x^2)+x$ на интервале $[−1,1]$, используя 250 точек. Не забудьте пометить оси!


```python
xx = np.linspace(-1, 1, 250)
yy = np.power(xx, 2) * np.sin(1 / np.power(xx, 2)) + xx
fig, ax = plt.subplots()
ax.plot(xx, yy, label='y=y(x)', color='#9966CC', linewidth=2)
plt.xlim(-1, 1)
plt.ylim(-1, 1.5)
ax.set_xlabel('x', fontsize=15)
ax.set_ylabel('y', fontsize=15)
ax.set_title('$x^2⋅sin(1/x^2)+x$', fontsize=15)
ax.legend()
```

2. Создайте semilogy график (с логарифмической шкалой для оси y и линейной шкалой для оси x) относительной разницы  функций $1/(1+x^2)$ и $1/x^2$ на интервале $[5,25]$. (Относительная разница чисел $a$ и $b$ определяется как $|1−a/b|$)

```python
xx = np.linspace(5, 25, 250)
yy = abs(1 - np.power(xx, 2) / (1 + np.power(xx, 2)))
fig, ax = plt.subplots(figsize=(5, 5))
ax.semilogy(xx, yy, linestyle='-', color='#00BFFF',  label='y=y(x)', linewidth='3', alpha=0.66)
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)
ax.set_title('разница функций $1/(1+x^2)$ и $1/x^2$')
ax.legend()
```

3. В Scipy также есть много менее распространенных специальных функций. Взгляните на обширный список <a href=https://docs.scipy.org/doc/scipy/reference/special.html> здесь</a>! Выберите несколько интересных спец функций, которые вы знаете (или, может быть, не знаете), подставьте несколько чисел и начертите их.

```python
xx = np.linspace(-10, 10, 225)
yy = scipy.special.gammasgn(xx)
fig, ax = plt.subplots(figsize=(5,5))

ax.plot(xx, yy, color='#FF1493', linestyle='--', linewidth=3, label='y=y(x)')
plt.xlim(-10, 5)
plt.ylim(-5, 5)
ax.set_xlabel('x', fontsize=15)
ax.set_ylabel('y', fontsize=15)
ax.set_title('gammasgn')
ax.legend(bbox_to_anchor=(1.02, 1))
```

## Exercise 2  
1. Построить график функции $sin(x)/x$ на отрезке $[-6\pi;6\pi]$ и добавить тики по оси $x$ с шагом $\pi/2$ от $-6\pi$ до $6\pi$, причём обозначения тиков должны содержать греческую букву $\pi$. Добавить сетку на график.

```python
xx = np.linspace(- 6 * np.pi, 6 * np.pi, 250)
yy = np.sin(xx) / xx
fig, ax = plt.subplots(figsize=(15, 3))

ticks = np.linspace(- 6 * np.pi, 6 * np.pi, 25)
tickslabels = []

for i in range(25):
    x = -6 * np.pi + i * np.pi / 2
    if x != 0:
        tickslabels.append(f'{i / 2 - 6}$\pi$')
    else:
        tickslabels.append(0)
        
ax.set_xticks(ticks)
ax.set_xticklabels(tickslabels)
x_minor_ticker = mpl.ticker.FixedLocator(np.linspace(- 11 * np.pi / 2, 11 * np.pi / 2, 12))
ax.xaxis.set_minor_locator(x_minor_ticker)
ax.set_xlabel('x', fontsize=15)
ax.set_ylabel('y', fontsize=15)
ax.set_title('$sin(x)/x$')
ax.grid()
ax.plot(xx, yy, label='y=y(x)', color='#FFA500', alpha=0.8, linewidth=3, linestyle='-')
ax.legend()
```
